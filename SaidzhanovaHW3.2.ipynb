```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
arr = np.loadtxt("http://www.scipy-lectures.org/_downloads/populations.txt")
```

```python
plt.plot(arr[:, 0], arr[:, 1], label = "Hares")
plt.plot(arr[:, 0], arr[:, 2], label = "Lynoxes")
plt.plot(arr[:, 0], arr[:, 3], label = "Carrots")
plt.xticks(np.arange(arr[:,0].min(), arr[:,0].max(),3))
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend()
plt.show()
```

```python
total = arr[:, 1:3].sum(axis = 1)
plt.figure(1, figsize=(8, 5))
plt.bar(arr[:,0], arr[:, 1]/total, bottom = arr[:,2]/total, label = "Hares")
plt.bar(arr[:,0], arr[:, 2]/total,  label = "Lynxes")
plt.legend()
plt.show()
```

```python
plt.hist(arr[:, 1], bins = 10, normed = True, color="pink")
plt.axvline(arr[:, 1].mean())
plt.show()
```

```python
plt.plot(arr[:, 0], arr[:, 1], label = "Hares")
plt.plot(arr[:, 0], arr[:, 2], label = "Lynoxes")
plt.plot(arr[:, 0], arr[:, 3], label = "Carrots")
plt.xticks(np.arange(arr[:,0].min(), arr[:,0].max(),3))
plt.xlabel("Time")
plt.ylabel("Population")
plt.scatter(arr[:,0][np.argmax(arr[:,1])],max(arr[:,1]))
plt.scatter(arr[:,0][np.argmax(arr[:,2])],max(arr[:,2]))
plt.scatter(arr[:,0][np.argmax(arr[:,3])],max(arr[:,3]))
plt.show()
```

```python
largest = np.where(arr[:, 1] > arr[:, 2], "H", "L")
print (largest)
plt.pie ([largest.tolist().count("H"), largest.tolist().count("L")], colors=["red","blue"])
plt.show()
```

```python
plt.plot(arr[:, 0], arr[:, 1], label = "Hares")
plt.plot(arr[:, 0], arr[:, 2], label = "Lynxes")
plt.plot(arr[:, 0], arr[:, 3], label = "Carrots")
plt.xticks(np.arange(arr[:,0].min(), arr[:,0].max(),3))
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend()
plt.scatter(arr[:,0][arr[:,1]>50000], arr[:, 1][arr[:,1]>50000], c="red")
plt.scatter(arr[:,0][arr[:,2]>50000], arr[:, 2][arr[:,2]>50000], c="red")
plt.scatter(arr[:,0][arr[:,3]>50000], arr[:, 3][arr[:,3]>50000], c="red")
plt.show()
```

```python
hares = np.sort(arr[:, 1])
lynxes = np.sort(arr[:,2])
carrots = np.sort(arr[:, 3])
print("hares", arr[:,0][arr[:,1] == hares[0]], arr[:,0][arr[:,1] == hares[1]])
print("lynxes", arr[:,0][arr[:,2] == lynxes[0]], arr[:,0][arr[:,2] == lynxes[1]])
print("carrots", arr[:,0][arr[:,3] == carrots[0]], arr[:,0][arr[:,3] == carrots[1]])
```

```python
newhares = [0] + [arr[:,1][i]-arr[:,1][i-1] for i in range(1 ,len(arr[:, 1]))]
newlynxes = [0] + [arr[:,2][i]-arr[:,2][i-1] for i in range(1 ,len(arr[:, 2]))]
plt.plot(arr[:,0], newhares, label = "Hares change")
plt.plot(arr[:, 0],newlynxes, label = "Lynxes change")
plt.xlabel("Time")
plt.legend()
plt.show()
eh = np.mean(arr[:,1])
el = np.mean(arr[:,2])
ehs = np.mean(arr[:,1]*arr[:,1])
els = np.mean(arr[:,2]*arr[:,2])
ehl = np.mean(arr[:,1]*arr[:,2])
corr = (ehl - eh*el)/pow((ehs-eh*eh)*(els-el*el),1/2) 
print("Corr =", corr)
```
